```python
#Get duration 
helloWorld_total_frames = helloWorld.getnframes()
helloWorld_time = helloWorld_total_frames/helloWorld_Framerate
helloWorld_total_frames
```

# Comparing 2 Sound Waves

```python
#====================================
# Processing Audio File "goodMorning"
#====================================
#Import Audio file as wave object
goodMorning = wave.open("Sample/goodMorning.wav",'r')
#Convert wave object to bytes
goodMorning_byte = goodMorning.readframes(-1)
#Convert Sound Wave bytes to integers
goodMorning_int = np.frombuffer(goodMorning_byte, dtype='int16')
#Get wave object frequency
goodMorning_freq = goodMorning.getframerate()
#Get time duration of the Audio file
goodMorning_totalFrames = goodMorning.getnframes()
goodMorning_time = goodMorning_totalFrames/goodMorning_freq
#Get Timestamps
goodMorning_ts = np.linspace(start=0, 
                             stop=len(goodMorning_int)/goodMorning_freq, 
                             num=len(goodMorning_int))

#======================================
# Processing Audio File "goodAfternoon"
#======================================
#Import Audio file as wave object
goodAfternoon = wave.open("Sample/goodAfternoon.wav",'r')
#Convert wave object to bytes
goodAfternoon_byte = goodAfternoon.readframes(-1)
#Convert Sound Wave bytes to integers
goodAfternoon_int = np.frombuffer(goodAfternoon_byte, dtype='int16')
#Get wave object frequency
goodAfternoon_freq = goodAfternoon.getframerate()
#Get time duration of the Audio file
goodAfternoon_totalFrames = goodAfternoon.getnframes()
goodAfternoon_time = goodAfternoon_totalFrames/goodAfternoon_freq
#Get Timestamps
goodAfternoon_ts = np.linspace(start=0, 
                             stop=len(goodAfternoon_int)/goodAfternoon_freq, 
                             num=len(goodAfternoon_int))


print(f"goodMorning frequency\n {goodMorning_freq}")
print(f"goodAfternoon frequency\n {goodAfternoon_freq}")
print(f"\ngoodMorning timestamp\n {goodMorning_ts}")
print(f"goodAfternoon timestamp\n {goodAfternoon_ts}")
print(f"\ngoodMorning integer values\n {goodMorning_int}")
print(f"goodAfternoon integer values\n {goodAfternoon_int}")
```

```python
#======================================================
# Visualizing Audio Files "goodMorning & goodAfternoon"
#======================================================
#Initialize figure and setup title
plt.title("goodMorning VS goodAfternoon")
#x & y labels
plt.xlabel("time (s)")
plt.ylabel("amplitude")
#Add timestamp and  values
plt.plot(goodMorning_ts, goodMorning_int, label="goodMorning")
plt.plot(goodAfternoon_ts, goodAfternoon_int, label="goodAfternoon", alpha =0.5)
#create legend and show
plt.legend()
plt.show()
```
